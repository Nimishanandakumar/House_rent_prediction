```python
df = df.drop(['Point of Contact','Posted On','Area Locality'],axis=1)
```

```python
df.columns
```

```python
col = (['BHK', 'Size', 'Floor', 'Area Type', 'City','Furnishing Status', 'Tenant Preferred', 'Bathroom','Rent'])
```

```python
df = df[col]
df
```

```python
df['Floor'] = df['Floor'].apply(lambda x: x.split()[0] if isinstance(x,str) else x)
```

```python
df['Floor'].unique()
```

```python
replace_dict = {'Ground':'0','Upper':'-1','Lower':'-2'}
df['Floor']=df['Floor'].replace(replace_dict,regex=True)
df
```

```python
df.to_csv('preprocessed_data.csv')
```
